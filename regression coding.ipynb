```python
#GRADIENT DESCENT  LINEAR
## Step 1: Define the model

import torch.nn as nn

class LinearModel(nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.theta_0 = nn.Parameter(torch.tensor([1.]))
        self.theta_1 = nn.Parameter(torch.tensor([1.]))
    
    def forward(self, x):
        '''
        x can be a minibatch of scalars.
        '''
        return (self.theta_0 + self.theta_1 * x)

linear_model = LinearModel()
```

```python
# Step 2: Define the loss function

class SquareLoss(nn.Module):
    def __init__(self):
        super(SquareLoss, self).__init__()
    
    def forward(self, inp, target):
        d = (inp - target) ** 2
        return torch.mean(d)

squared_loss = SquareLoss()

# Note: You can get squared loss (and a lot of other conventional loss functions) as a predefined loss in pytorch.
```

```python
# Step 3: Define the optimizer: SGD (Stochastic Gradient Descent)

optimizer = torch.optim.SGD(linear_model.parameters(), lr=0.01)
```

```python
# Step 4: Defining the training function

def train(model, x, y, optimizer, loss_fn):
    n_epochs = 1000
    theta_0s = []
    theta_1s = []
    losses = []

    for epoch in range(n_epochs):
        # zero the parameter gradients
        optimizer.zero_grad()
        
        op = model(x)
        loss = loss_fn(op, y)
        
        loss.backward()
        optimizer.step()
        
        theta_0s.append(model.theta_0.item())
        theta_1s.append(model.theta_1.item())
        losses.append(loss.item())
        print (f'==> Epoch {epoch+1}, loss: {loss.item()}, theta_0: {model.theta_0.item()}, theta_1: {model.theta_1.item()}')
        
    return theta_0s, theta_1s, losses
```

```python
# Step 5: Training the model

theta_0s, theta_1s, losses = train(linear_model, x_linear, y_linear, optimizer, squared_loss)
'''when lr is too large, the values change too dramatically and we get nan after a while. the right numbers are found earlier, 
but we keep moving away from them. if the number is too small, we move too slowly and we end up not getting close
enough to what we want'''
# Do the parameters theta_0 and theta_1 look like the ones we used for generating the data at the end of training?
# What happens when we play around with the learning rate for the optimizer? What if we set it to 0.001? What if we set it to 0.1?
```

```python
class PolyModel(nn.Module):
    def __init__(self):
        super(PolyModel, self).__init__()
        self.theta_0 = nn.Parameter(torch.tensor([1.]))
        self.theta_1 = nn.Parameter(torch.tensor([1.]))
        self.theta_2 = nn.Parameter(torch.tensor([1.]))
        self.theta_3 = nn.Parameter(torch.tensor([1.]))
    
    def forward(self, x):
        '''
        x can be a minibatch of scalars.
        '''
        return (self.theta_0 + self.theta_1 * (x**3) + self.theta_2*(x**2) + self.theta_3*x)

polynomial_model = PolyModel()
```

```python
class SquareLoss(nn.Module):
    def __init__(self):
        super(SquareLoss, self).__init__()
    
    def forward(self, inp, target):
        d = (inp - target) ** 2
        return torch.mean(d)

squared_loss = SquareLoss()
```

```python
optimizer = torch.optim.Adam(polynomial_model.parameters(), lr=0.01) #adam changes the learning rate as it goes (adaptive learning rate)
```

```python
def train(model, x, y, optimizer, loss_fn):
    n_epochs = 1000
    theta_0s = []
    theta_1s = []
    theta_2s = []
    theta_3s = []
    losses = []

    for epoch in range(n_epochs):
        # zero the parameter gradients
        optimizer.zero_grad()
        
        op = model(x)
        loss = loss_fn(op, y)
        
        loss.backward()
        optimizer.step()
        
        theta_0s.append(model.theta_0.item())
        theta_1s.append(model.theta_1.item())
        theta_2s.append(model.theta_2.item())
        theta_3s.append(model.theta_3.item())
        losses.append(loss.item())
        print (f'==> Epoch {epoch+1}, loss: {loss.item()}, theta_0: {model.theta_0.item()}, theta_1: {model.theta_1.item()}, theta_2: {model.theta_2.item()}, theta_3: {model.theta_3.item()}')
        
    return theta_0s, theta_1s, theta_2s, theta_3s, losses
```

```python
theta_0s, theta_1s, theta_2s, theta_3s, losses = train(polynomial_model, x_poly, y_poly, optimizer, squared_loss)
```
